```python
# %% [markdown]
# #### Breaking book corpus in sentences  and then preprocessing it

# %%
import os
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
import nltk

nltk.download('punkt')

story = []
for filename in os.listdir('data'):
    file_path = os.path.join('data', filename)
    with open(file_path, 'r', encoding='latin1') as f:  # Replace 'latin1' with the correct encoding if known
        corpus = f.read()
        raw_sent = sent_tokenize(corpus)
        # print(raw_sent)
        for sent in raw_sent:
            story.append(simple_preprocess(sent))

```

```python

# %%
def find_dimensions(lst):
    outer_dim = len(lst)
    inner_dim = [len(sublist) for sublist in lst] if lst else []
    return outer_dim, inner_dim

# Example list of lists

dimensions = find_dimensions(story)

print("Outer Dimension (Number of sublists):", dimensions[0])
print("Inner Dimensions (Lengths of sublists):", dimensions[1])


```

```python
# %% [markdown]
# #### Building the Word2Vec model using Gensim

# %%
model = gensim.models.Word2Vec(
    window=10, # Prediction of Centre word taking 10 -10 words both the sides
    #Eg:- window = 1  "Hi how are you all good?"
    # """1. Hi how are
    #     2. how are you
    #     3. ae you all
    #     4.you all good"""
    min_count=2, # Choose only those sentences which 2 words atleast
    workers=4   
        )
```

```python

# %% [markdown]
# #### Building my Vocalulary i.e., all the unique words in my CORPUS

# %%
model.build_vocab(story)
```

```python

# %%

# Get the number of words in the vocabulary
vocab_size = len(model.wv.key_to_index)

print("Number of words in the vocabulary:", vocab_size)
```

```python

# %%
model.corpus_count
```

```python

# %%
# DEfaukt epoch = 5

model.train(story,
            total_examples = model.corpus_count, # No of sentences in the whole corpus
            epochs = model.epochs) # Default epochs = 5
```
